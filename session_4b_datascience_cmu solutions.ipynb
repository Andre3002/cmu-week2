## Basic Matplotlib

Creating the most basic plot

```python
# Example of a simple line plot
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create a basic plot
plt.plot(x, y)

# Show the plot
plt.show()
```

Customizing the plot axis, title, legend, and grid

```python
# Example of a simple line plot
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create a basic plot
plt.plot(x, y)

# Customize the plot
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(["Line 1"], 
           loc='upper left')
plt.grid(False)

# Show the plot
plt.show()
```

Customizing the color, line styles and markers

```python
# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create a basic plot
plt.plot(x, y, marker='o', linestyle='--', color='r')

# other markers: 'o', 's', '^', 'x', 'D', 'p'
# other linestyles: '-', '--', '-.', ':'
# other colors: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'

# Show the plot
plt.show()
```

Customizing ticks and using different scales

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y = np.exp(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Exponential Growth")

# Customizing ticks
# X-axis ticks from 0 to 10 with step 1
plt.xticks(np.arange(0, 11, step=1))
# Custom Y-axis tick values
plt.yticks([0, 5000, 25000, 75000, 200000])
# Setting log scale for y-axis
plt.yscale('log')

# Show the plot
plt.show()
```

Figure and Axes

```python
# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Creating a figure and axes
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Plot with Subplots")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

# Show the plot
plt.show()
```

Multiple plots in the same figure

```python
# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Two subplots in one figure
fig, (ax1, ax2) = plt.subplots(1, 2)  # 1 row, 2 columns
# change the figure size
fig.set_size_inches(10, 5)

ax1.plot(x, y)
ax1.set_title("Plot 1")
ax2.scatter(x, y)
ax2.set_title("Plot 2")
plt.show()
```

## Ploting with Pandas

Pandas leverages Matplotlib to create quick, basic visualizations directly from Pandas DataFrames or Series.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Bar Chart: Comparing total bill across different days
total_bill_sum = df.groupby('day')['total_bill'].sum()
print(total_bill_sum)

plt.figure(figsize=(8, 6))
total_bill_sum.plot(kind='bar', color='skyblue')
plt.title("Total Bill by Day (Sum)")
plt.xlabel("Day")
plt.ylabel("Total Bill")

plt.show()
```

```python
# Line Plot: Trends of total bill over table size
plt.figure(figsize=(8, 6))
bill_by_size_mean = df.groupby('size')['total_bill'].mean()
print(bill_by_size_mean)

bill_by_size_mean.plot(kind='line', color='blue', marker='o')
plt.title("Average Total Bill by Table Size")
plt.xlabel("Table Size")
plt.ylabel("Average Total Bill")
plt.show()
```

```python
# Scatter Plot: Relationship between total bill and tip
plt.figure(figsize=(8, 6))
df.plot(kind='scatter', x='total_bill', y='tip', color='purple')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Histogram: Distribution of total bills
plt.figure(figsize=(8, 6))
df['total_bill'].plot(kind='hist', bins=10, 
                      color='orange', edgecolor='black')
plt.title("Distribution of Total Bills")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot: Distribution of total bills by day
plt.figure(figsize=(8, 6))
df.boxplot(column='total_bill', by='day', grid=False)
plt.title("Box Plot of Total Bill by Day")
plt.suptitle('')  # Suppress automatic title
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Pie Chart: Proportion of meals served by day
plt.figure(figsize=(8, 6))
df['day'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Proportion of Meals Served by Day")
plt.ylabel('')  # Hide y-label
plt.show()
```

```python
# Correlation between numerical features
plt.figure(figsize=(8, 6))
correlation = df.corr(numeric_only=True)
print(correlation)

plt.imshow(correlation, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(correlation)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation)), correlation.columns)
plt.title("Correlation Heatmap")
plt.show()
```

## Ploting with Matplotlib

```python
# Bar Chart: Comparing total bill across different days
plt.figure(figsize=(8, 6))
df.groupby('day')['total_bill'].sum().plot(kind='bar', color='skyblue')
plt.title("Total Bill by Day (Sum)")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Line Plot: Trends of total bill over table size
df_line = df.groupby('size')['total_bill'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.plot(df_line['size'], df_line['total_bill'], 
         marker='o', color='blue')
plt.title("Average Total Bill by Table Size")
plt.xlabel("Table Size")
plt.ylabel("Average Total Bill")
plt.grid(True)
plt.show()
```

```python
# Scatter Plot: Relationship between total bill and tip
plt.figure(figsize=(8, 6))
plt.scatter(df['total_bill'], df['tip'], c='purple', alpha=0.6)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.grid(True)
plt.show()
```

```python
# Histogram: Distribution of total bills
plt.figure(figsize=(8, 6))
plt.hist(df['total_bill'], bins=10, color='orange', edgecolor='black')
plt.title("Distribution of Total Bills")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Box Plot: Distribution of total bills by day
plt.figure(figsize=(8, 6))
df.boxplot(column='total_bill', by='day', grid=False)
plt.title("Box Plot of Total Bill by Day")
plt.suptitle('')  # Suppress automatic title
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Pie Chart: Proportion of meals served by day
day_counts = df['day'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%')
plt.title("Proportion of Meals Served by Day")
plt.show()
```

```python

# Correlation between numerical features
correlation = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
plt.imshow(correlation, cmap='coolwarm', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(correlation)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation)), correlation.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Violin Plot: Distribution of total bills by day
plt.figure(figsize=(8, 6))

# Extract unique days dynamically from the dataset
days = df['day'].unique()

# Prepare data for violin plot, grouped by day
data = [df[df['day'] == day]['total_bill'] for day in days]

# Create violin plot
plt.violinplot(data, showmeans=True, showmedians=True)

# Customize the plot
plt.title("Violin Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.xticks(range(1, len(days) + 1), days)
plt.show()
```

## Ploting with seaborn

Seaborn is a Python data visualization library built on top of Matplotlib, designed to make it easier to create informative plots. Provides high-level functions for creating a wide variety of plots.

And it generally looks nicer than matplotlib defaults :)

```python
import seaborn as sns
import numpy as np
```

```python
# Bar Chart: Comparing total bill across different days
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", data=df, 
            estimator=np.sum, color='skyblue', ci=None)
plt.title("Total Bill by Day (Sum)")
plt.show()
```

```python
# Line Plot: Trends of total bill over table size
plt.figure(figsize=(8, 6))
sns.lineplot(x="size", y="total_bill", data=df, ci=None, 
             color='blue', marker='o')
plt.title("Total Bill by Table Size (Trend)")
plt.xlabel("Table Size")
plt.ylabel("Average Total Bill")
plt.show()
```

```python
# Scatter Plot: Relationship between total bill and tip
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", data=df, color='purple')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Histogram: Distribution of total bills
plt.figure(figsize=(8, 6))
sns.histplot(df['total_bill'], bins="auto", kde=True)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
# Box Plot: Distribution of total bills by day
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
# Pie Chart: Proportion of meals served by day
plt.figure(figsize=(8, 6))
df['day'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Proportion of Meals Served by Day")
plt.ylabel('')  # Hide y-label
plt.show()
```

```python
# Correlation between numerical features
plt.figure(figsize=(8, 6))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Violin Plot: Distribution of total bills by day
plt.figure(figsize=(8, 6))
sns.violinplot(x="day", y="total_bill", data=df, color='skyblue')

# Customize the plot
plt.title("Violin Plot of Total Bill by Day")
plt.xlabel("day")
plt.ylabel("total_bill")
plt.show()
```

## Plotly
Plotly is a powerful, interactive data visualization library for Python that allows for the creation of plots. 

```python
# Import necessary libraries
import plotly.express as px   # to install run pip install plotly
import plotly.graph_objs as go
import pandas as pd
```

```python
# Bar Chart: Comparing total bill across different days
fig1 = px.bar(df, x='day', y='total_bill', color='day', 
              title="Total Bill by Day (Sum)", 
              labels={'total_bill': 'Total Bill'})
fig1.show()
```

```python
# Line Plot: Trends of total bill over table size
df_line = df.groupby('size')['total_bill'].mean().reset_index()
fig2 = px.line(df_line, x='size', y='total_bill', 
               title="Average Total Bill by Table Size", 
               labels={'size': 'Table Size', 'total_bill': 'Average Total Bill'})
fig2.show()
```

```python
# Scatter Plot: Relationship between total bill and tip
fig3 = px.scatter(df, x='total_bill', y='tip', 
                  title="Scatter Plot of Total Bill vs Tip", 
                  labels={'total_bill': 'Total Bill', 'tip': 'Tip'},
                  color='day')
fig3.show()
```

```python
# Histogram: Distribution of total bills
fig5 = px.histogram(df, x='total_bill', nbins=10, 
                    title="Distribution of Total Bills", 
                    labels={'total_bill': 'Total Bill'})
fig5.show()
```

```python
# Box Plot: Distribution of total bills by day
fig6 = px.box(df, x='day', y='total_bill', 
              title="Box Plot of Total Bill by Day", 
              labels={'total_bill': 'Total Bill', 'day': 'Day'})
fig6.show()
```

```python
# Pie Chart: Proportion of meals served by day
day_counts = df['day'].value_counts().reset_index()
day_counts.columns = ['day', 'count']
fig7 = px.pie(day_counts, names='day', values='count', 
              title="Proportion of Meals Served by Day", 
              color_discrete_sequence=px.colors.sequential.RdBu)
fig7.show()
```

```python
# Correlation between numerical features
correlation = df.corr(numeric_only=True).round(2)  # Calculate correlation
fig4 = go.Figure(data=go.Heatmap(
    z=correlation.values,
    x=correlation.columns,
    y=correlation.columns,
    colorscale='viridis'))
fig4.update_layout(title="Correlation Heatmap")
fig4.show()
```

```python
# Violin Plot: Distribution of total bills by day
fig9 = px.violin(df, x='day', y='total_bill', 
                 box=True, points="all", 
                 title="Violin Plot of Total Bill by Day", 
                 labels={'total_bill': 'Total Bill', 'day': 'Day'})
fig9.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# load the data
dataset_path = "datasets/1- mental-illnesses-prevalence.csv"
ex_df = pd.read_csv(dataset_path)

# lets simplify the column names
new_collumn_names = {
    'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized': 'Schizophrenia',
    'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized': 'Depression',
    'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized': 'Anxiety',
    'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized': 'Bipolar',
    'Eating disorders (share of population) - Sex: Both - Age: Age-standardized': 'Eating',
}

# rename the columns based on the new_collumn_names dictionary
ex_df = ex_df.rename(columns=new_collumn_names)
```

### 1. Line Plot of Mental Disorder Trends Over Time

Plot the trend of depressive disorders globally over the years (using a line plot).

Hint: use groupby()

Also customize the plot with the following:
- title - "Trend of Depressive Disorders Globally Over the Years"
- x-axis "Year"
- y-axis -"Depressive Disorders (%)"

```python
# Plot the trend of depressive disorders globally over the years (using a line plot):

# 1. group the data by year and do the mean
depressive_disorders = ex_df.groupby('Year')['Depression'].mean()

# 2. create a line plot
plt.figure(figsize=(10, 6))
depressive_disorders.plot(kind='line', marker='o', color='b')

# 3. Customize the plot
plt.title("Trend of Depressive Disorders Globally Over the Years")
plt.xlabel("Year")
plt.ylabel("Depressive Disorders (%)")
plt.show()
```

```python
# seaborn line plot
# Plot the trend of depressive disorders globally over the years (using a line plot):

# 1. group the data by year and do the mean
depressive_disorders = ex_df.groupby('Year')['Depression'].mean().reset_index()

# 2. create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Depression', data=depressive_disorders, marker='o', color='b')
```

### 2. Bar Plot of Mental Disorder Prevalence by Country

Create a bar plot to show the average prevalence of Depression disorders for the top 10 countries with the highest prevalence.

Hint: use groupby and sorting operations

```python
# 1. group the data by Entity and do the mean
depressive_disorders = ex_df.groupby('Entity')['Depression'].mean()

# 2. sort the values
depressive_disorders = depressive_disorders.sort_values(ascending=False)

# 3. get the top 10 values
top_10_depressive_disorders = depressive_disorders.head(10)

# 4. create a bar plot
plt.figure(figsize=(10, 6))
top_10_depressive_disorders.plot(kind='bar', color='skyblue', title='Top 10 Countries with the Highest Depression Rates')
plt.xlabel('Country')
plt.ylabel('Depression (%)')
plt.show()
```

```python
# seaborn bar plot
# 1. group the data by Entity and do the mean
depressive_disorders = ex_df.groupby('Entity')['Depression'].mean().reset_index()

# 2. sort the values
depressive_disorders = depressive_disorders.sort_values(by='Depression', ascending=False)

# 3. get the top 10 values
top_10_depressive_disorders = depressive_disorders.head(10)

# 4. create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Entity', y='Depression', data=top_10_depressive_disorders, color='skyblue')
plt.title('Top 10 Countries with the Highest Depression Rates')
plt.xlabel('Country')
plt.ylabel('Depression (%)')
plt.xticks(rotation=90)
plt.show()
```

### 3. Scatter Plot between Schizophrenia and Bipolar Disorders

Create a scatter plot to show the relation between Schizophrenia and Bipolar disorders.
For the top-10 countries in terms of Bipolar disorder.

```python
# Create a scatter plot to show the relation between Schizophrenia and Bipolar disorders. For the top-10 countries in terms of Bipolar disorder.

# 1. get the top 10 countries with the highest bipolar disorder rates
top_10_bipolar_disorders = ex_df.groupby('Entity')['Bipolar'].mean().sort_values(ascending=False).head(10)

# 2. filter the data to get only the top 10 countries
top_10_countries_data = ex_df[ex_df['Entity'].isin(top_10_bipolar_disorders.index)]

# 3. create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(top_10_countries_data['Schizophrenia'], top_10_countries_data['Bipolar'])
plt.title("Relation between Schizophrenia and Bipolar Disorders")
plt.xlabel("Schizophrenia (%)")
plt.ylabel("Bipolar (%)")
plt.show()
```

```python
# seaborn
# Create a scatter plot to show the relation between Schizophrenia and Bipolar disorders. For the top-10 countries in terms of Bipolar disorder.

# 1. get the top 10 countries with the highest bipolar disorder rates
top_10_bipolar_disorders = ex_df.groupby('Entity')['Bipolar'].mean().sort_values(ascending=False).head(10)

# 2. filter the data to get only the top 10 countries
top_10_countries_data = ex_df[ex_df['Entity'].isin(top_10_bipolar_disorders.index)]

# 3. create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Schizophrenia', y='Bipolar', data=top_10_countries_data)
plt.title("Relation between Schizophrenia and Bipolar Disorders")
plt.xlabel("Schizophrenia (%)")
plt.ylabel("Bipolar (%)")
plt.show()
```

### 5. Pie Chart for Mental Disorder Distribution in a Specific Country

Create a pie chart to visualize the distribution of different mental disorders in a specific country (e.g., "Portugal") for a given year (e.g., 2010):

```python
# 1. Filter the data for Portugal in 2010
country_2010_data = ex_df[(ex_df['Entity'] == 'Portugal') & (ex_df['Year'] == 2010)]

# 2. Get only the disorder columns
disorder_columns = ex_df.columns[3:]

# 3. Get the first and only row of the filtered data
country_disorders = country_2010_data[disorder_columns].iloc[0]

# 3. Plot the pie chart
country_disorders.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Mental Disorder Distribution in the United States (2010)')
plt.ylabel('')
plt.show()
```
